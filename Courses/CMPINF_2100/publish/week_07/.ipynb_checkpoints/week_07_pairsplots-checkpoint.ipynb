### Visualizing relationships between continuous variables via PAIRS PLOTS

The previous recordings focused mostly on just 2 continuous variables. However, we saw that the CORRPLOT scale to MANY pairs of continuous variables. The PAIRS PLOTS is a way to visualize the relationships between MULTIPLE variables in one figure!

## Import Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

## READ DATA

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.info()
```

## SCATTER PLOT

Visualizes the relationship between two continuous variables.

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm")

plt.show()
```

```python
sns.relplot(data=penguins, x="body_mass_g", y="bill_length_mm")

plt.show()
```

```python
penguins.info()
```

We still have not looked at all pairs of continous variables!!

This might seem like a very tedious task..to make the scatter plot between all pairs!!

## PAIRS PLOTS

It will make all pairs of scatterplots for you!!

The function is the sns.pairplot() function. It is a FIGURE LEVEL function because it generates MANY facets!!!

```python
sns.pairplot(data=penguins)

plt.show()
```

Confirm one of the marginal histograms!!

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="hist", bins=10)

plt.show()
```

We can GROUP oy CONDITION on cat variables within PAIRS PLOTS just like we did with SCATTER PLOTS and TREND PLOTS!!

We CAN COLOR by categorical variables via the hue argument!!!

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="species")

plt.show()
```

```python
sns.lmplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="species")

plt.show()
```

We can color PAIRS PLOTS via hue!!

```python
sns.pairplot(data=penguins, hue="species")

plt.show()
```

The PAIR PLOTS are incorrect above!

The DEFAULT pairsplot in Seaborn is WRONG because the sample size is infuencing the kde chart. We must change the common_norm=False.

The MARGINAL KDE is shown below for one varibale.

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="kde")

plt.show()
```

The conditional KDE is shown with the WRONG settings!

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="kde", hue="species")

plt.show()
```

Remove the sample size effect with common_norm=False.

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="kde", hue="species", common_norm=False)

plt.show()
```

The CONDITIONAL KDE should give you the same conclusiong as the VIOLIN PLOT!!

```python
sns.catplot(data=penguins, x="species", y="flipper_length_mm", kind="violin", inner="quartile")

plt.show()
```

You must remove the sample size effect with the PAIRS PLOT just as you did with the CONDITIONAL KDE!!!

```python
sns.pairplot(data=penguins, hue="species",
            diag_kws={"common_norm":False})

plt.show()
```

The pairs plot works best when you have less than 10 variables.

You can get away with up to 15-20 but anything more than 20 is INFORMATION OVERLOAD!!!

Because 20 variables...produces a PAIRS PLOT with a 20x20 plot matrix!!!
