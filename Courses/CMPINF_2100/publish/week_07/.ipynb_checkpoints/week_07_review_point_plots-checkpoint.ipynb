## Point Plot
The point plot shows or visualizes the **CONDITIONAL MEAN** of a continuous var **GIVEN** a CAT variable.

That means it is CATEGORICAL to CONTINUOUS relationship plot.

In Seaborn, it is part of the sns.catplot() FIGURE LEVEL function family because it is FOCUSED on categorical groupings!!

Usually, the cat var is associated with the x-axis and the cont with the y-axis.

The CONDITIONAL mean of the lifeExp GIVEN the year. 

```python
sns.catplot(data=gap_df, x="year", y="lifeExp", kind="point", aspect=2, join=False)

plt.show()
```

We could have grouped by the continent variable.

```python
sns.catplot(data=gap_df, x="continent", y="lifeExp", kind="point", aspect=2, join=False)

plt.show()
```

But we could also GROUP BY 2 cat vars!!

One or the PRIMARY categorical variable is associated with the x-axis and the SECONDARY categorical variable is associated with the marker hue or color.

```python
sns.catplot(data=gap_df, x="year", y="lifeExp", kind="point", hue="continent", aspect=2, join=False)

plt.show()
```

## What is the vertical line??

The vertical line is an ERROR BAR.

The ERROR BAR is showing us the 95% CI on the AVG!!!

The CONFIDENCE INTERVAL is denoted by CI.

```python
sns.catplot(data=gap_df, x="year", y="lifeExp", kind="point", aspect=2, join=False, errorbar=("ci", 95))

plt.show()
```

We can change the CI size for example we can show the 68% CI.

```python
sns.catplot(data=gap_df, x="year", y="lifeExp", kind="point", aspect=2, 
            errorbar=("ci", 68), join=False)

plt.show()
```

What is the CI? And **HOW** is the CI calculated??

The CI comes from the SEM!!!!

The CI is a multiple of the SEM around the AVERAGE!!!

Begin by GROUPING BY the cats and SUMMARIZING the continuous!!

```python
gap_df.groupby(['year']).\
aggregate(lifeExp_count = ("lifeExp", "count"),
          lifeExp_avg   = ("lifeExp", "mean"),
          lifeExp_sem   = ("lifeExp", "sem")).\
reset_index()
```

Lets assign our GROUPED and SUMMARIZED DataFrame to an object!

```python
gap_year_summary = gap_df.groupby(['year']).\
aggregate(lifeExp_count = ("lifeExp", "count"),
          lifeExp_avg   = ("lifeExp", "mean"),
          lifeExp_sem   = ("lifeExp", "sem")).\
reset_index()
```

```python
gap_year_summary
```

LETS CALCULATE THE 95% CI USING THE CONDITIONAL MEANS and their respective SEM.

The LOWER BOUND on the 95% CI is the MEAN - 2*SEM.

THe UPPER BOUND on the 95% CI is the MEAN + 2*SEM.

These formulas come from the fact that the distribution on the MEAN is a Gaussian according to the CLT!!!!

```python
gap_year_summary['lifeExp_ci_lwr'] = gap_year_summary.lifeExp_avg - 2*gap_year_summary.lifeExp_sem

gap_year_summary['lifeExp_ci_upr'] = gap_year_summary.lifeExp_avg + 2*gap_year_summary.lifeExp_sem
```

```python
gap_year_summary
```

Return to the POINT PLOT.

```python
sns.catplot(data=gap_df, x="year", y="lifeExp", kind="point", aspect=2, join=False)

plt.show()
```

Use the AXIS LEVEL function for greater control.

```python
fig, ax = plt.subplots(figsize=(12,6))

sns.pointplot(data=gap_df, x="year", y="lifeExp", join=False, ax=ax)

plt.show()
```

We can examine the ATTRIBUTES or METHODS for the AXIS object.

```python
ax.get_xlim()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

sns.pointplot(data=gap_df, x="year", y="lifeExp", join=False, ax=ax)

ax.hlines(y=[47, 51.1], xmin=ax.get_xlim()[0], xmax=ax.get_xlim()[1], color="r")

plt.show()
```

```python
gap_year_summary
```

In Summary, the POINT PLOT performs the following actions for us:
* Calculates the CONDITIONAL MEAN (AVERAGE) for the y-axis variable GIVEN each x-axis CATEGORY
* Calculates the SEM for the y-axis variable GIVEN each x-axis CATEGORY
* Calculates the LOWER and UPPER BOUND of the CI
* Plots the CONDITIONAL MEAN as a POINT
* Plots the CI as an ERROR BAR between the LOWER AND UPPER BOUNDs
