## Continuous to Continuous Relationships

The fundamental plot for showing the relationship between 2 continuous vars is the **SCATTER PLOT**.

In Seaborn, the FIGURE LEVEL FUNCTION is the sns.relplot() function.

The x-axis and the y-axis are both associated with continuous variables.

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm")

plt.show()
```

```python
sns.relplot(data=penguins, x="bill_depth_mm", y="flipper_length_mm")

plt.show()
```

```python
penguins.info()
```

```python
sns.relplot(data=penguins, x="body_mass_g", y="bill_length_mm")

plt.show()
```

The underlying AXIS LEVEL FUNCTIOn called by the FIGURE LEVEL sns.replot() is sns.scatterplot() function.

```python
fig, ax = plt.subplots()

sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", ax=ax)

plt.show()
```

In Summary, the SCATTER PLOT shows the relationship between two continuous variables.

They can show more than 2 variables!!

A very common practice is to INCLUDE a categorical variable via color!! We can associate a cat var with the hue argument

Lets return to our first plot!!

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm")

plt.show()
```

We have categorical variables, object data types, in the penguins dataset!

```python
penguins.info()
```

What if we GROUP BY a categorical variable? In other words, we can study if the RELATIONSHIP between the 2 variables CHANGES across categories!!!

Lets assign the species variable to the hue argument in sns.replot().

```python
sns.relplot(data=penguins, x="body_mass_g", y="bill_length_mm", hue="species")

plt.show()
```

Instead of coloring by the categorical variable, we can also use FACETS to separate the groups!!

```python
sns.relplot(data=penguins, x="body_mass_g", y="bill_length_mm", col="species")

plt.show()
```

You can combine hue and col facets to get facets and colors.

```python
sns.relplot(data=penguins, x="body_mass_g", y="bill_length_mm", hue="species", col="species")

plt.show()
```

HOWEVER...if you do NOT color by the same variable as the FACET...that means you can use DIFFERENT categorical variables for color and facet!!!

You can group the relationship on two separate and distinct cat vars. You are grouping by the combo of the two cat vars!

```python
sns.relplot(data=penguins, x="body_mass_g", y="bill_length_mm", col="species", hue="island")

plt.show()
```

We can add yet another categorical variable through ROW facets!!

```python
penguins.info()
```

If we use ROW and COL facts and associate hue with a cat variable we are effectively GROUPING the relationship between 2 continuous variables on the combo of 3 categorical vars!!!

```python
sns.relplot(data=penguins, x="body_mass_g", 
                           y="bill_length_mm", 
                           col="island", 
                           row="species", 
                           hue="sex")

plt.show()
```

Remember to always check the COMBOS between cats before looking at relationships between continuous vars!!

```python
sns.catplot(data=penguins, x="island", hue="species", kind="count")

plt.show()
```
