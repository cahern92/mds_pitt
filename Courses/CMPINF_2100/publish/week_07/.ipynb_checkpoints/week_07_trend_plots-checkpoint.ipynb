### Visualize continuous to continuous relationships via TREND PLOTS

The TREND PLOT is an extension of the SCATTER PLOT that adds further quantitative info about the relationship between continuous variables.
## Import Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

## READ DATA

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.info()
```

## Scatter plot

Shows the relationship between two continuous variables.

```python
sns.relplot(data=penguins,
            x="flipper_length_mm",
            y="bill_length_mm")

plt.show()
```

We saw that by GROUPING or CONDITIONING on a cat variable...our interpretation of the scatterplot changed.

```python
sns.relplot(data=penguins,
            x="flipper_length_mm",
            y="bill_length_mm",
           hue="species")

plt.show()
```

## Trend Plots

We can add trend lines to visualize the **LINEAR** relationship between two continuous variables.

TREND LINES correspond to the **BEST FIT LINE** between two variables.

Later on..we will learn HOW this line is created..but for now lets visualize it and let the line or TREND HELP US interpret the relationship!

The TREND is not crated with the sns.relplot() function.

Instead, the TREND is created with the FIGURE LEVEL function...sns.lmplot().

```python
sns.lmplot(data=penguins,
           x="flipper_length_mm",
           y="bill_depth_mm")
plt.show()
```

We are viewing an ANTI or NEGATIVE relationship between the two variables!!

BUT what happens if we GROUP OR CONDITION on a cat variable!!

Lets COLOR using the hue argument by species.

```python
sns.lmplot(data=penguins,
           x="flipper_length_mm",
           y="bill_depth_mm",
          hue="species")
plt.show()
```

GIVEN OR GROUPING BY OR CONDITIONING ON the species CHANGES our opinion about the RELATIONSHIP between the 2 continuous variables! GROUPING BY species reveals a POSITIVE relationship between the two!!!

We can of course create the TREND PLOT between 2 different continuous variables!
