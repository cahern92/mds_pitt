## Boxplot

The boxplot is a tool for exploring the relationships between CATEGORICAL and CONTINUOUS variables.

In Seaborn, it is part of the sns.catplot() FIGURE LEVEL function.

The x variable is associated with the categorical variable and the y variable is associated with the continuous variable.

The BOXPLOT summarizes the CONDITIONAL DISTRIBUTION with SEVERAL summary statistics.

```python
sns.catplot(data=gap_df, x="year", y="lifeExp", kind="box", aspect=2)

plt.show()
```

Lets look at the relationship between continent and lifeExp.

```python
sns.catplot(data=gap_df, x="continent", y="lifeExp", kind="box", aspect=2)

plt.show()
```

In Summary, Bixplots allow us to answer several IMPORTANT questions.

* Does the MEDIAN appear to change across categories?
* Does the VARIATION appear to change across the categories?
* More generally, are the summary statistics different across the categories?

## What is the boxplot best suited for?

The boxplot is at its best when the CONDITIONAL DISTRIBUTION are roughly SYMMETRIC. 

We must check the SHAPE of the conditional distributions!!

One way to do that is with the FACETED histograms!!

Lets return to the CONDITIONAL DISTRIBUTION of lifeExp GIVEN year.

```python
sns.displot(data=gap_df, x="lifeExp", col="year", kind="hist", col_wrap=4)

plt.show()
```

None of the facets show EXTREME lack of symmetry!!

Lack of symmetry is referred to as a SKEW.

What does EXTREME SKEW look like? And what are the consequences of EXTREME skew??

Load in a dataset that has EXTREME SKEW!!

```python
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds.info()
```

```python
diamonds.dtypes
```

```python
diamonds.count()
```

The cat data type is a SPECIAL data type for CATEGORICAL vars within Pandas.

```python
diamonds.color.value_counts(normalize=True)
```

```python
sns.catplot(data=diamonds, x="color", kind="count", aspect=2)

plt.show()
```

Lets now examine the MARGINAL DISTRIBUTION of the continuous variable price.

```python
sns.displot(data=diamonds, x="price", kind="hist", bins=25, aspect=2)

plt.show()
```

Lets include the KDE on top of the histogram to get a SMOOTHER picture of the distribution.

```python
sns.displot(data=diamonds, x="price", kind="hist", bins=25, aspect=2, kde=True)

plt.show()
```

Lets now examine the CONDITIONAL DISTRIBUTION of price GIVEN color.

We will visualize the categories via facets.

By default, ALL FACETS SHARE the x and y-axis bounds!!

Sometimes we need to MODIFY the bounds to allow them to be NOT SHARED or FREE!!

Lets change the y-axis bounds to be FREE in the conditional distribution figure.

```python
sns.displot(data=diamonds, x="price", col="color", kind="hist", bins=25, kde=True,
           facet_kws={"sharey":False}, col_wrap=2)

plt.show()
```

What is the consequence of the SKEW or LONG TAIL behavior on the BOXPLOT??

```python
sns.catplot(data=diamonds, x="color", y="price", kind="box", aspect=2)

plt.show()
```

But are the most frequent intervals contained in the BOXES???

EXTREMELY SKEWED distributions will have the MOST FREQUENT INTERVALS near the LOWER or UPPER bounds!!

The boxplot CANNOT represent theat behavior because BOXPLOTS cannot show the distsribution shape!!

We can see the shape with FACETED HISTOGRAMS but a draw back of FACETS is we have to look at many subplots!!

What if we used CONDITIONAL KDE plots with the cat var associated with the hue argument???

```python
sns.displot(data=diamonds, x="price", hue="color", kind="kde", common_norm=False, aspect=2)

plt.show()
```

CONDITIONAL KDE PLOTS can be very busy to look at when we have more than 4 categories!!

## Violin Plots

Shows CONDITIONAL KDE plots side by side within a single axis rather than across multiple subplots!!

```python
sns.catplot(data=diamonds, x="color", y="price", kind="violin", aspect=2)

plt.show()
```

```python
sns.displot(data=diamonds, x="price", col="color", kind="kde",
           facet_kws={"sharey": False})

plt.show()
```

We can show HORIZONTAL lines for the QUANTILES

```python
sns.catplot(data=diamonds, x="color", y="price", kind="violin", aspect=2,
           inner="quartile")

plt.show()
```

In Summary:

* Always look at boxplots to study how the summary statistics CHANGE across categories.
* But..you must ALSO look at if the distributional SHAPE is SYMMETRIC and if the distributional shape CHANGES across categories!
    * Use faceted histograms/kds plots
    * Use violin plots
