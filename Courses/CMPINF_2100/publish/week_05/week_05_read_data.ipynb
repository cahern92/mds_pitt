## Headers or Column Names
By default, the `pd.read_` family of functions ASSUMES the top row is the HEADER ROW!!!

The top row therefore does NOT contain values!! Instead, is it assumed that the TOP ROW or HEADER ROW contains the columns NAMES!

```python
df0.columns
```

```python
df0.dtypes
```

But let's see what happens if we work with a data set or a SHEET within an EXCEL workbook that does NOT use a HEADER ROW!

When you know there is NO HEADER, then the `header` argument must be set to `None`.

```python
df0_no_names = pd.read_excel("Example_Excel_Data.xlsx", sheet_name="no_headers", header=None)
```

```python
df0_no_names.info()
```

```python
df0_no_names
```

If there is no header row, we can use the `names` argument to NAME the columns!

```python
df0_no_names_2 = pd.read_excel("Example_Excel_Data.xlsx",
                               sheet_name="no_headers",
                               header=None,
                               names=df0.columns)
```

```python
df0_no_names_2
```

```python
df0_no_names_2.info()
```

But...be VERY careful...if there is a HEADER row and you specify that there isn't by mistake!!!

```python
df0_mistake = pd.read_excel("Example_Excel_Data.xlsx", header=None)
```

```python
df0_mistake
```

## Read CSV

CSV files have the extension `.csv`. Unlike Excel workbooks, they contain a single spreadsheet rather than multiple sheets.

```python
pd.read_csv("Example_A.csv")
```

`pd.read_csv()` has nearly all the same arguments as `pd.read_excel()`.

```python
pd.read_csv("Example_A.csv", index_col=0)
```

But a few other important argument taht you may need.

We can specify the MAX NUMBER of rows to read.

```python
pd.read_csv("Example_A.csv", nrows=3)
```

```python
pd.read_csv("Example_A.csv", skiprows=2)
```

When you skip rows, be very careful with the HEADER!!!

```python
pd.read_csv("Example_A.csv", skiprows=2, header=None)
```
