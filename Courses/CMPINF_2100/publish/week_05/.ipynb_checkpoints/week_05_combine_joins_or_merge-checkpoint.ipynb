## JOIN or MERGE

The `pd.merge()` function JOINS two DFs.

We need to specify the TWO DFs AND we need to specify the column (or columns) that the two are JOINED ON!!!

The JOINING variable is referred to as the **KEY**!!!

```python
pd.merge(dfA, dfB, on="E")
```

```python
pd.merge(dfA, dfB, on="E").info()
```

How many rows existed in `dfA`?

```python
dfA.shape
```

What happened to the 3 missing rows in `dfA`?

The answer has to do with the UNIQUE values of the KEY between the LEFT and the RIGHT data sets!!!

```python
dfA
```

```python
dfB
```

```python
pd.merge(dfA, dfB, on="E")
```

The `E == 'cc'` value is NOT present in the JOINED dataset!!!

This is because in addition to specifying the KEY that JOIN ON, we must also specify **HOW** the LEFT and RIGHT DFs are JOINED!!!

By default Pandas uses the **INNER JOIN**

The INNER JOIN keeps the values that are shared between the two datasets!!!

```python
pd.merge(dfA, dfB, on="E", how="inner")
```

There are other types of JOINS!!

The LEFT JOIN keeps all KEY values present in LEFT dataset.

We do NOT know the value of `G` associated with `E == 'cc'`. That is why those rows are MISSING in the JOINED data set!

```python
pd.merge(dfA, dfB, on="E", how="left")
```

The NaN standas for NOT A NUMBER!!

The NaN means the value is MISSING!!!

If we would instead use a RIGHT JOIN, then all the KEY VALUES present in the RIGHT data set are retained!!

```python
pd.merge(dfA, dfB, on="E", how="right")
```

So far this example, the RIGHT JOIN and INNER JOIN produce the same result!

The LEFT JOIN generated MISSING values.

**HOWEVER** that may not always be the case!!

You might have situations where the INNER JOIN behaves like the LEFT JOIN and the RIGHT JOIN generates the MISSING!!!

Lets see what happens if we use a RIGHT JOIN between `dfA` and `dfC`.

```python
dfA
```

```python
dfC
```

```python
pd.merge(dfA, dfC, on="F", how="right")
```

```python
pd.merge(dfA, dfC, on="F", how="left")
```

```python
pd.merge(dfA, dfC, on="F", how="inner")
```

MISSING Values can be generated by JOINS!!

We are bringing together OCLUMNS from one data set into another data set!

If we do not have an exact association between the columns, then missings occur.

I personally like to use the LARGER dataset as the LEFT and the smaller as the RIGHT.

Also, `pd.merge()` function is available as a method via the `.merge()` method for Pandas DataFrames.

```python
dfA.merge(dfC, on="F", how="right")
```

The `merge()` method allows us to CHAIN TOGETHER multiple JOINS in a single line of code.

Lets join the `dfA` and `dfB` and then join that result to `dfC`.

```python
pd.merge(dfA, dfB, on="E", how="left").merge(dfC, on="F", how="right")
```

I am on purpose trying to generate the max amount of MISSINGs. The INNER JOIN used by Pandas by default tries to minimize MISSINGs.

Lets assign the result of this JOIN to a new variable.

```python
df = pd.merge(dfA, dfB, on="E", how="left").merge(dfC, on="F", how="right")
```

```python
df.sort_values("B", inplace=True, ignore_index=True)
```
