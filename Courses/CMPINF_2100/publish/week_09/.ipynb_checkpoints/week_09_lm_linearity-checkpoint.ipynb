### What does LINEAR actually mean?
We use the phrase LINEAR MODEL or LINEAR REGRESSION all the time...but what does that actually mean????
## Import Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

## What do you think LINEAR means?
When most people hear LINEAR REGRESSION they think "best fit lines". They think a straight line between output and the input.

But now you know the MODEL is actually predicting the AVERAGE OUTPUT!!!

So does LINEAR mean the AVERAGE OUTPUT is LINEARLY RELATED to the INPUT???

The math notation for the conventional LINEAR MODEL:

$$\mu = \beta_0 + \beta_1 \times x $$

But..what if we instead **PREPROCESSED** OR TRANSFORMED the input through a NON-LINEAR transformation function.

For example, let's define a NEW COLUMN `s` which is equal to the SIN(X) or the SINE FUNCTION applied to the input X.

The math notation looks like:

$$\mu = \beta_0 + \beta_1 \times s $$

The FEATURE depends on the INPUT!!! Or, the FEATURE is a FUNCTION of the input!!!

$$\mu = \beta_0 + \beta_1 \times sin(x) $$

Formally, the DEFINITION of LINEAR has **NOTHING** to do with the RELATIONSHIP between the TREND (AVERAGE OUTPUT) and the INPUT!!!

LINEAR means the TREND is LINEARLY related to the UNKNOWN regression coefficients!!

Lets use simulated data to demonstrate all of the ASSUMPTIONS of the linear model just as we did in the previous example.

## Make data
To make the data, we need to specify the coefficients and the sigma of the linear model.

I am picking these values but they could be different.

```python
b0_sin = 0.25
```

```python
b1_sin = -2.25
```

```python
sigma_sin = 0.1
```

Generate input values from negative PI to positive PI.

```python
df_sin = pd.DataFrame({"x": np.linspace(-3.14159, 3.14159, num=25)})
```

```python
df_sin.head()
```

The TREND or AVERAGE OUTPUT is a function of the INPUT!!!

```python
df_sin["trend"] = b0_sin + b1_sin * np.sin(df_sin.x)
```

```python
sns.relplot(data=df_sin, x="x", y="trend", kind="line")

plt.show()
```

Obs of the output are NORMALLY distributed **GIVEN** the TREND!!

If we know the TREND we know everything about the distribution ASSUMED by the linear model!!!

Use random number generator to GENERATE the random outputs!!!

```python
rg = np.random.default_rng(2100)
```

```python
df_sin["y"] = rg.normal(loc=df_sin.trend.to_numpy(), 
                        scale=sigma_sin,
                        size=df_sin.shape[0])
```

```python
sns.relplot(data=df_sin, x="x", y="y")

plt.show()
```

The above data were generated from a LINEAR MODEL that satisfies the **ALL** assumptions of the linear model!!!

To see that this is indeed a linear model, lets create the FEATURE `s` in the DataFrame!

```python
df_sin["s"] = np.sin(df_sin.x)
```

Now visualize the RELATIONSHIP between the random OUTPUT and the FEATURE derived from the input!!

```python
sns.relplot(data=df_sin, x="s", y="y")

plt.show()
```

## Simulation
Lets use the same simulation procedure as the previous example to reinforce the ASSUMPTIONS of the linear model!!

We need a function that calculates the TREND for a given INPUT and INTERCEPT and SLOPE. When then randomly generates OUTPUTS around  the TREND for a given SIGMA value.

```python
def generate_lm_sin(rid, xuse, b0, b1, sigma_use):
    res_df = pd.DataFrame({"x": xuse})
    res_df["trend"] = b0 + b1 * np.sin(res_df.x)


    res_df["y"] = rg.normal(loc=res_df.trend.to_numpy(),
                            scale=sigma_use,
                            size=res_df.shape[0])
    res_df["rep_id"] = rid
    return res_df
```

```python
generate_lm_sin(0, df_sin.x, b0_sin, b1_sin, sigma_sin).info()
```

Replicate the process 9 times.

```python
study_9_list = []

for ix in range(9):
    study_9_list.append(generate_lm_sin(ix, df_sin.x, b0_sin, b1_sin, sigma_sin))
```

```python
study_9_df = pd.concat(study_9_list, ignore_index=True)
```

```python
study_9_df
```

Visualize the replications via a facets!!!

```python
sns.relplot(data=study_9_df, x="x", y="y", col="rep_id", col_wrap=3)

plt.show()
```

We can SUMMARIZE by calculating the **AVERAGE** output at each INPUT location!! 

```python
sns.catplot(data=study_9_df, x="x", y="y", kind="point", aspect=2)

plt.show()
```

Replicate the process 500 times!!

```python
study_500_list = []

for ix in range(500):
    study_500_list.append(generate_lm_sin(ix, df_sin.x, b0_sin, b1_sin, sigma_sin))
```

```python
study_500_df = pd.concat(study_500_list, ignore_index=True)
```

```python
study_500_df.head()
```

Visualize the CONDITIONAL DISTRIBUTION of `y` **GIVEN** `x`.

```python
sns.catplot(data=study_500_df, x="x", y="y", kind="violin", aspect=2)

plt.show()
```

## Why does this matter??
Far too often people will think that a linear model or LINEAR REGRESSION only applies if they see a LINEAR TREND!!

But that is NOT true!!

You do NOT need to jump straight to NEURAL NETWORKS, DEEP LEARNERS, or Gradient Boosted Trees!!

Very often, you just need to derive FEATURES from your INPUTS that are NON-LINEAR functions!! You can FIT linear models given those new FEATURES!!!!!
