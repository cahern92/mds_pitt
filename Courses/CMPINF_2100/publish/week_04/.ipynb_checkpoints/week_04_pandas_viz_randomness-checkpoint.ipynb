### Use Pandas to help visualize the random simulation

We will use Pandas methods to visualize the random numbers generated in our simulation and their summary statistics.

## Import Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Generate random numbers

Let's start with creating seven random numbers from a normal distribution

We need to initialize the random number generator object.

```python
rg = np.random.default_rng(2100)
```

Lets replicate generating 7 random numbers 25 times.

Assign the data to a 2D numpy array.

```python
x = rg.random((7,25))
```

```python
x.shape
```

Lets convert the NumPy Array into a Pandas DataFrame.

```python
xdf = pd.DataFrame(x)
```

```python
xdf.info()
```

```python
xdf
```

```python
xdf.describe()
```

```python
x.mean(axis=0)
```

```python
xdf.describe().index
```

```python
xdf.describe().loc[["mean", "std"]]
```

## Visualization

Boxplot can help visualize the CENTRAL behavior AND the variation of COLUMNS in DataFrames!

Let's use a boxplot now to understand the variability in the samples across the REPLICATIONS.

Please remember, the COLUMNS in `xdf` are REPLICATIONS of generating 7 random numbers!

```python
fig, ax = plt.subplots(figsize=(12,6))
xdf.boxplot(ax=ax)
plt.show
```

But the horiznal green line is NOT the AVERAGE.

Lets include the AVG in the boxplot!

```python
fig, ax = plt.subplots(figsize=(12,6))
xdf.boxplot(ax=ax, showmeans=True)
plt.show
```

Let's change the green triangle to a red triangle.

```python
fig, ax = plt.subplots(figsize=(12,6))
xdf.boxplot(ax=ax, showmeans=True, meanprops={"markerfacecolor": "red", "markeredgecolor":"red"})
plt.show()
```

But we use far more than just 25 replications in our simulations.

We saw that when thousands of reps were used the SIMULATION bheaved like theory.

The CLT showed the reps thousands of times produced a GAUSSIAN or BELL CURVE distribution for the avgs!

Let's now use 500 reps to see the variation in the sample avg based on 7 random samples.

```python
rg = np.random.default_rng(2100)
```

```python
xdf_more_reps = pd.DataFrame(rg.random((7, 500)))
```

```python
xdf_more_reps
```

```python
xdf_more_reps.shape
```

```python
fig, ax = plt.subplots(figsize=(12,6))
xdf_more_reps.boxplot(ax=ax, showmeans=True, meanprops={"markerfacecolor": "red", "markeredgecolor":"red"})
plt.show()
```
