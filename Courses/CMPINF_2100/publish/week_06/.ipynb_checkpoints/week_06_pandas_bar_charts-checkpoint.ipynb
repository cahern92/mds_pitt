Lets now visualize the counts with a BAR CHART!!

Although matplotlib will manage the figure..we will NOT call a matplotlib function to generate the plot!!

Lets use Pandas plotting methods to make the bar chart!

We must still being by DEFINING the object oriented style figure and axes objects!!

```python
fig, ax = plt.subplots()

gap_df.continent.value_counts().plot(kind="bar", ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

gap_df.continent.value_counts().plot(kind="bar", ax=ax)

plt.show()
```

The text can be made easier to read using a HORIZONTAL bar chart!

```python
fig, ax = plt.subplots(figsize=(12,6))

gap_df.continent.value_counts().plot(kind="barh", ax=ax)

plt.show()
```

Sometimes you want to show the PROPORTION rather than the count!

```python
gap_df.continent.value_counts(normalize=True)
```

A prop bar chart is created by applying the .plot() method to tbe normalized value_counts() result!

```python
fig, ax = plt.subplots(figsize=(12, 6))

gap_df.continent.value_counts(normalize=True).plot(kind="barh", ax=ax)

plt.show()
```

You must visualize the barc chart for every single categorical variable!!

But you must explore every variable!!!

## Penguins dataset

```python
penguins.dtypes
```

```python
penguins.nunique()
```

```python
penguins.sex.value_counts(normalize=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

penguins.sex.value_counts().plot(kind="bar", ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

penguins.sex.value_counts(normalize=True).plot(kind="bar", ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

penguins.species.value_counts().plot(kind="barh", ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

penguins.species.value_counts(dropna=False).plot(kind="barh", ax=ax)

plt.show()
```

```python
penguins.isna().sum()
```

Lets make bar chart for sex var and dont remove missings!

```python
fig, ax = plt.subplots(figsize=(8, 4))

penguins.sex.value_counts(dropna=False).plot(kind="barh", ax=ax)

plt.show()
```

Lets look at the island variable!

```python
fig, ax = plt.subplots(figsize=(8, 4))

penguins.island.value_counts(dropna=False).plot(kind="barh", ax=ax)

plt.show()
```
